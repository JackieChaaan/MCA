```python
import pandas as pd
from numpy.matrixlib.defmatrix import matrix
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import seaborn as sns
from pandas.plotting import scatter_matrix
```

```python
df1=pd.read_csv("iris.csv")
df1.head()
```

```python
df1.tail()
```

```python
#Histogram
plt.figure(figsize=(10,8))
sns.histplot(df1['sepal.length'],bins=20,kde=True)
plt.title("Histogram")
plt.xlabel('Sepal length(cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box Plot
df1.plot(kind='box')
plt.show()
```

```python
#Quartile Box(Box Plot)
plt.figure(figsize=(8,6))
sns.boxplot(x='variety',y='sepal.length',data=df1)
plt.xlabel('variety')
plt.ylabel('sepal.length(cm)')
plt.title("Quartile(Box Plot)")
plt.show()
```

```python
#Distribution Chart
plt.figure(figsize=(10,8))
sns.kdeplot(data=df1,x='sepal.width',hue='variety',fill='True')
plt.xlabel('sepal.width')
plt.ylabel('density')
plt.title("Distribution Chart")
plt.show()
```

```python
#Scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=df1,x='petal.length',y='petal.width',hue='variety',size='sepal.length')
plt.xlabel('petal length(cm)')
plt.ylabel('petal width(cm)')
plt.title("Scatter Plot")
plt.show()
```

```python
#Scatter Matrix
sns.set(style='ticks')
sns.pairplot(df1,hue='variety',markers=('o','s','d'))
plt.suptitle('Scatter matrix pairplot',y=1.02)
plt.show()
```

```python
#Bubble Chart
plt.figure(figsize=(10,6))
sns.scatterplot(data=df1,x='sepal.length',y='sepal.width',hue='variety',size='sepal.length')
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width(cm)')
plt.title("Bubble Chart")
plt.show()
```

```python
#Density Chart
plt.figure(figsize=(10,6))
sns.kdeplot(data=df1,x=df1['petal.length'],y=df1['petal.width'],hue='variety',cmap="Blues",fill=True,cbar=True)
plt.xlabel('petal length(cm)')
plt.ylabel('petal width(cm)')
plt.title("Density Chart")
plt.show()
```

```python
#Deviation Chart
import matplotlib.pyplot as plt

# Calculate the mean sepal length
mean_sepal_length = df1['sepal.length'].mean()

# Calculate deviation from the mean sepal length
df1['deviation'] = df1['sepal.length'] - mean_sepal_length

plt.figure(figsize=(10, 6))
plt.bar(df1['variety'], df1['deviation'], color='skyblue')
plt.xlabel('Species/Variety')
plt.ylabel('Deviation from mean sepal length')
plt.title('Deviation Chart')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python
#Andrews Curve
from pandas.plotting import andrews_curves
plt.figure(figsize=(10,6))
andrews_curves(df1,'variety',colormap='viridis')
plt.title('Andrews curves for IRIS')
plt.show()
```
