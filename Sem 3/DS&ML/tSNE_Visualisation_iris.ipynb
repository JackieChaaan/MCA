#**Importing Necessary Packages**

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

#**Dataset Loading**

```python
# Load the Iris dataset
data = pd.read_csv('iris.csv')
```

```python
# Separate features and labels
X = data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y = data['variety']
```

```python
# Map the variety (class) labels to unique colors
color_mapping = {'Setosa': 'r', 'Versicolor': 'g', 'Virginica': 'b'}
colors = [color_mapping[variety] for variety in y]
```

#**2D Visualization**

```python
# Perform t-SNE for 2D visualization
tsne = TSNE(n_components=2, random_state=0)
X_2d = tsne.fit_transform(X)
```

```python
# Create a 2D scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=colors)
plt.title('t-SNE Visualization (2D)')
plt.show()
```

#**3D Visualization**

```python
# Perform t-SNE for 3D visualization
tsne = TSNE(n_components=3, random_state=0)
X_3d = tsne.fit_transform(X)
```

```python
# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_3d[:, 0], X_3d[:, 1], X_3d[:, 2], c=colors)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('t-SNE Visualization (3D)')
plt.show()
```
