```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data=pd.read_csv("iris.csv")
print(data)
```

```python
print(data.shape)
print(data.info())

print(data.isnull().sum())

data.boxplot()
data.hist()
plt.show()
```

```python
x=data.drop('variety',axis=1)
y=data['variety']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
dcl=DecisionTreeClassifier()
dcl.fit(x_train,y_train)

predict=dcl.predict(x_test)
accuracy=accuracy_score(y_test,predict)
```

```python
print(f"Accuracy:{accuracy:.2f}")
cm=confusion_matrix(y_test,predict)
print("Confussion Matrix")
print(cm)
print(data.columns)
print(data['variety'].unique())


feature_names = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

# Class names
class_names = data['variety'].unique()

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(dcl, feature_names=feature_names, class_names=class_names, filled=True)
plt.show()
```
